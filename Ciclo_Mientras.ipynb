 **Introducción**:




Un ciclo while ejecuta instrucciones repetidamente, mientras que una condición definida en el ciclo sea verdadera.

Una estructura clásica del ciclo while se compone de las siguientes partes:
while1.png


1. Se inicializa (declara) la variable de control.

```python
i=1
```

2. En la línea siguiente se utiliza la palabra reservada de Python while.
3. Se prueba la variable de control (se define una condición). La cual puede ir entre paréntesis (son opcionales).
4. Se agregan dos puntos (**:**).

```python
while i <= 5:
```

5. En la línea siguiente de forma tabulada se define el cuerpo del while. Allí se coloca la instrucción (o instrucciones) que se ejecutarían si la condición se evalúa a True. 


```python
  print(i)#imprimimos el valor actual de la variable de control
```

6. Antes de finalizar el cuerpo del while se realiza un cambio sobre la variable de control.

```python
  i+=1#incrementamos el valor de la variable de control
```

Veamos ahora el programa completo:

```python
i=1#Declaramos einicializamos la varaibal que controlará el ciclo
while i <= 5:# probamos la variable de control mediante una condicion que nos permitirá saber en qué momneto finalizará nuestro programa
  print(i)#imprimimos el valor de la variable de contol
  i+=1#incrementamos la variable control en cada iteración
```
